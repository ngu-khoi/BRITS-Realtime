```python
#Subcell loads all data into dataframes
import os
import re
import pandas as pd
patient_ids = []

for filename in os.listdir('./raw/set-a'):
    # the patient data in PhysioNet contains 6-digits
    match = re.search('\d{6}', filename)
    if match:
        id_ = match.group()
        patient_ids.append(id_)
        
dataframe_collection = {}
counter = 0
for id_ in patient_ids:
    dataframe_collection[counter] = pd.read_csv("./imputed_correlation/"+ id_ +".csv", encoding='unicode_escape')
    counter += 1
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
#Each dataframe can be accessed by dataframe_collection[x]
#Each correlation matrix is split by 12/11
#This makes many correlations invisible. This however is the only way to observe attributes with correlations matrices due to unreadability for large attributes
#Modify x to look at various patients [0, 3996]
x = 0
df = pd.DataFrame({'DiasABP':dataframe_collection[x]['DiasABP'], 'HR':dataframe_collection[x]['HR'], 'Na':dataframe_collection[x]['Na'], 'Lactate':dataframe_collection[x]['Lactate'], 'NIDiasABP':dataframe_collection[x]['NIDiasABP'], 'PaO2':dataframe_collection[x]['PaO2'], 'WBC':dataframe_collection[x]['WBC'], 'pH':dataframe_collection[x]['pH'], 'Albumin':dataframe_collection[x]['Albumin'], 'ALT':dataframe_collection[x]['ALT'], 'Glucose':dataframe_collection[x]['Glucose'], 'SaO2':dataframe_collection[x]['SaO2']})
corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```
